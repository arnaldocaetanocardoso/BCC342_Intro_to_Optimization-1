```python
import line_search as ls
import math
import matplotlib.pyplot as plt
import numpy as np
import fmin
```

```python
# f = lambda x: x[0]**2 + x[1]**2 + x[0]*x[1]
# grad = lambda x: [2*x[0] + x[1], 2*x[1] + x[0]]
# H = lambda x: [[2,1],[1,2]]

f = lambda x: (100*(x[1] - x[0]**2)**2 + (x[0]-1)**2) 
grad = lambda x: [-400*(x[1]-x[0]**2)*x[0] + 2*(x[0]-1), 200*(x[1]-x[0]**2)]
H = lambda x: [[-400*x[1]+1200*x[0]**2,-400*x[0]],[-400*x[0],200]]

delta = 0.025
x = np.arange(-3.0, 3.0, delta)
y = np.arange(-2.0, 2.0, delta)
X, Y = np.meshgrid(x, y)
Z = np.ndarray((len(X),len(X[0])))

for i in range(len(X)):
    for j in range(len(X[0])):
        Z[i][j] = f([X[i][j],Y[i][j]])

```

```python
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z, 200)
```

```python
xs,fs,xlist,fxlist = fmin.gradient_descent([1,-2],f,grad,tol=1e-10,alpha=[])

print(xs,fs)
```

```python
x0 = [x[0] for x in xlist]
x1 = [x[1] for x in xlist]

fig, ax = plt.subplots()
ax.contour(X, Y, Z, 50)
ax.plot(x0,x1,'^b-')
ax.plot(xs[0],xs[1],'or')
```

```python
plt.plot(fxlist,'bo')
```

```python
xs,fs,xlist,fx_list = fmin.newton_method([1,-2],f,grad,H)

print(xs)
print(fs)
```

```python
x0 = [x[0] for x in xlist]
x1 = [x[1] for x in xlist]

fig, ax = plt.subplots()
ax.contour(X, Y, Z, 50)
ax.plot(x0,x1,'^b-')
ax.plot(xs[0],xs[1],'or')
```

```python
plt.plot(fx_list,'bo')
```
