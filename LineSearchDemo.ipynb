```python
import line_search as ls
import math
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = lambda x: 2 + math.cos(x) + math.cos(2*x -1/2)/2

xs = [x for x in np.arange(1,6,0.001)]
fxs = [f(x) for x in xs]
```

```python
plt.plot(xs,fxs)
```

```python
x_s = ls.golden_search(f,1,6,0.00001)
```

```python
plt.plot(xs,fxs)
plt.plot(x_s,f(x_s),'or')
```

```python
x_s = ls.quadratic_fit_search(f,1,6,0.00001)
plt.plot(xs,fxs)
plt.plot(x_s,f(x_s),'or')
```
